```python
# Langkah 3 - Data Imputation
# Age - mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Cabin - "DECK"
df['Cabin'] = df['Cabin'].fillna("DECK")

# Embarked - modus
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```
